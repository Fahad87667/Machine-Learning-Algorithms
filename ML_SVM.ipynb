```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification,make_blobs, make_moons
from sklearn.model_selection import train_test_split
```

```python
x1,y1 = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, class_sep=2.0, random_state=42)
```

```python
plt.scatter(x1[:,0],x1[:,1],c=y1)
plt.show()
```

```python
x2,y2 = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=32)
```

```python
plt.scatter(x2[:,0],x2[:,1],c=y2)
plt.show()
```

```python
x3,y3 = make_moons(n_samples=1000, noise=0.5, random_state=32)
```

```python
plt.scatter(x3[:,0],x3[:,1],c=y3)
plt.show()
```

```python
x_train1, x_test1, y_train1, y_test1= train_test_split(x1,y1, test_size=0.2, random_state=32)
```

```python
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification

clf = LinearSVC()
clf.fit(x_train1,y_train1)
```

```python
clf.score(x_train1,y_train1)
```

```python
y_pred1=clf.predict(x_test1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test1,y_pred1)
```

```python
x_train2, x_test2, y_train2, y_test2= train_test_split(x3,y3, test_size=0.2, random_state=32)
```

```python
from sklearn.svm import LinearSVC

clf = LinearSVC()
clf.fit(x_train2,y_train2)
```

```python
clf.score(x_train2,y_train2)
```

```python
y_pred2=clf.predict(x_test2)
```

```python
accuracy_score(y_pred2,y_test2)
```
