```python
df = pd.DataFrame({
    'iq': [110, 95, 105, 120, 88, 78, 92, 85, 60, 75,
           105, 98, 115, 82, 70, 68, 112, 99, 79, 89,
           94, 102, 118, 87, 73, 96, 110, 83, 74, 107,
           100, 116, 65, 91, 76, 108, 93, 77, 84, 97,
           72, 113, 81, 69, 121, 86, 71, 114, 101, 130],
    'cgpa': [7, 9, 8, 6, 5, 7, 8, 6, 4, 7,
             8, 8, 9, 6, 6, 7, 8, 7, 5, 6,
             8, 8, 9, 6, 6, 8, 7, 6, 5, 8,
             7, 9, 5, 7, 6, 8, 9, 7, 7, 8,
             6, 9, 7, 6, 9, 7, 6, 9, 7, 7],
    'salary': [7, 5, 8, 4, 6, 7, 9, 3, 5, 6,
               8, 7, 9, 4, 3, 6, 8, 5, 4, 7,
               9, 7, 9, 5, 4, 8, 6, 5, 3, 8,
               9, 3, 6, 5, 8, 7, 4, 6, 7, 9,
               8, 4, 5, 9, 6, 4, 9, 6, 5, 8]
})
```

```python
df.head()
```

```python
df['base_model'] = np.mean(df['salary'])
```

```python
df.head()
```

```python
df['residual']= df['salary']-df['base_model']
```

```python
df
```

```python
x=df[['iq','cgpa']]
```

```python
y = df['residual']
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
```

```python
dt.fit(x,y)
```

```python
dt.score(x,y)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
tree.plot_tree(dt,filled=True,max_depth=2,)
plt.show()
```
