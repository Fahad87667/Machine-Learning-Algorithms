```python
import pandas as pd
```

```python
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

```python
df['Price']=dataset.target
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#divide the dataset into independent and dependent
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# divide the dataset into train and test
from sklearn.model_selection import train_test_split

x_train , x_test, y_train, y_test = train_test_split(
    x,y, test_size=0.33, random_state=42
)
```

```python
x_train.shape,x_test.shape
```

```python
x.shape
```

```python
# 3. Feature Scaling -- Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
scaler.transform(x_train)
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
regression= LinearRegression()
regression.fit(x_train,y_train)
```

```python
regression.intercept_
```

```python
regression.coef_
```

```python
# prediction
y_predict = regression.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
mse = mean_squared_error(y_test,y_predict)
mae =mean_absolute_error(y_test,y_predict)
np.sqrt(mse)
```

```python
# Acuracy R-Squared and  adjusted R-Squared
from sklearn.metrics import r2_score
r2score=r2_score(y_test,y_predict)
r2score
```

```python
# Display adjusted R-squared
adjustedR2score = 1 - (1 - r2score) * (len(y) - 1) / (len(y) - x.shape[1] - 1)
adjustedR2score
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=20.0) # alpha parameter Ridge(alpha=)
ridge.fit(x_train,y_train)
```

```python
y_predict= ridge.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_predict)
mae =mean_absolute_error(y_test,y_predict)
print(np.sqrt(mse), mse, mae)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=20.0)
lasso.fit(x_train,y_train)
```

```python
y_predict = lasso.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_predict)
mae =mean_absolute_error(y_test,y_predict)
print(np.sqrt(mse), mse, mae)
```

```python
from sklearn.linear_model import ElasticNet
elasticnet = ElasticNet(alpha=20.0)
elasticnet.fit(x_train,y_train)
```

```python
y_predict = elasticnet.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_predict)
mae =mean_absolute_error(y_test,y_predict)
print(np.sqrt(mse), mse, mae)
```
